```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt

%matplotlib inline
```

### Load the Data Set

```python
breast_cancer = load_breast_cancer()
breast_cancer
```

### Create a DataFrame

```python
df = pd.DataFrame(breast_cancer.data,columns=breast_cancer.feature_names)
df['target'] = breast_cancer.target
df
```

```python
df.isnull()
```

```python
sns.heatmap(df.isnull(),yticklabels=False , cbar = False , cmap ='viridis' )
```

```python
sns.countplot(x = 'target', hue = 'target', data = df, palette='RdBu_r')
```

```python
cols = ['mean radius', 'mean texture', 'mean perimeter', 'mean smoothness', 'mean compactness', 'target']
sns.pairplot(df[cols], height = 2.5)
plt.show()
```

```python
X = df.drop('target', axis=1)
y = df['target'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
knn = KNeighborsClassifier(n_neighbors=5, metric='manhattan')
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
score= []
for i in range(1,40):
        knn = KNeighborsClassifier(i).fit(X_train, y_train)
        score.append(knn.score(X_train, y_train))

score
```

```python
plt.plot(np.arange(1,40), score)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
knn.score(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```
